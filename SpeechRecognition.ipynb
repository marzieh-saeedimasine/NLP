```python
import speech_recognition as sr
import matplotlib.pyplot as plt
from pydub import AudioSegment
```

```python
# Function to convert audio file to text
def audio_to_text(audio_file):
    recognizer = sr.Recognizer()
    audio = AudioSegment.from_wav(audio_file)

    with sr.AudioFile(audio_file) as source:
        audio_data = recognizer.record(source)

    try:
        text = recognizer.recognize_google(audio_data,language="sv-SE")
        return text
    except sr.UnknownValueError:
        return "Could not understand audio"
    except sr.RequestError as e:
        return f"Error with the speech recognition service; {e}"
```

```python
audio_file_path = "./Svenska-koder.wav"
text_result = audio_to_text(audio_file_path)
```

```python
# Plotting the audio waveform using Matplotlib
audio_data = AudioSegment.from_wav(audio_file_path)
samples = audio_data.get_array_of_samples()

plt.plot(samples)
plt.title("Audio Waveform")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.show()

print("Transcription:")
print(text_result)
```
